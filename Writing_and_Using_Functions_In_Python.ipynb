```python
def greet():
    print("Hello! Welcome to Python Functions.")


greet()
```

```python

def greet_user(name):
    print(f"Hello, {name}!")

greet_user("Neda")
greet_user("Abby")
```

```python

def greet_user(name="Guest"):
    print(f"Welcome, {name}!")

greet_user()
greet_user("Neda")
```

```python
def print_names(*names):
    print("Names passed:", names)
    for name in names:
        print("Hello", name)

print_names("Neda", "Bindhu", "Pranavi")
```

```python
def print_info(**info):
    print("User Info:")
    for key, value in info.items():
        print(f"{key} = {value}")

print_info(name="Neda", age=19, city="Karimnagar")
```

```python

def show_details(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

show_details(1, 2, 3, name="Neda", city="Hyderabad")
```

```python
def multiply(a, b):
    return a * b

def square(n):
    return multiply(n, n)

print("Square of 6:", square(6))
```
