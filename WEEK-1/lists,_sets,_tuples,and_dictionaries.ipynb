```python

fruits = ['apple', 'banana', 'mango', 'mango']
print("Original list:", fruits)


print("First fruit:", fruits[0])
print("Second fruit:",fruits[1])

```

```python
fruits.append('cherry')
print("After appending:", fruits)

```

```python
fruits.remove('banana')
print("After removing banana:", fruits)


print("Number of fruits:", len(fruits))
```

```python
dimensions = (5, 15, 25)
print("Tuple:", dimensions)

```

```python
print("First element:",dimensions[0])
print("Second elemeny:",dimensions[1])
```

```python
numbers={1,2,3,4,3,2,1}
print("Set(no duplicates):",numbers)
```

```python
numbers.add(5)
print("After adding:",numbers)
```

```python
numbers.remove(3)
print("After removing:",numbers)
```

```python
set_a={1,2,3}
set_b={3,4,5}
print("Union:",set_a | set_b)
print("Intersection:",set_a & set_b)
```

```python

student = {
    'name': 'Alice',
    'age': 19,
    'course': 'Python'
}
print("Dictionary:", student)

```

```python
print("Student name:", student['name'])
```

```python
student['grade'] = 'A'
print("After adding grade:", student)


del student['age']
print("After deleting :", student)

for key, value in student.items():
    print(f"{key}: {value}")
```

```python
keys = ['id', 'name', 'marks']
values = [2, 'Neda', 88]
student_info = dict(zip(keys, values))
print("Student Info from zip:", student_info)
```
